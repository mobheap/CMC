```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
column_names = ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT', 'MEDV']
data = pd.read_csv('housing.csv', delim_whitespace=True, names=column_names, index_col=None)
```

```python
data.head()
```

```python
data.to_csv('housingbetter.csv')
```

```python
plt.subplots(figsize=(20,10))
sns.set(style='whitegrid')
sns.boxplot(data)
```

```python
corr = data.corr()
plt.subplots(figsize=(20,10))
dataplot = sns.heatmap(corr, cmap='YlGnBu', annot=True)
```

```python
x = data.LSTAT.values
y = data.MEDV

plt.scatter(x, y)
plt.xlabel('LSTAT')
plt.ylabel('MEDV')
```

```python
x = x.reshape(-1, 1)

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
predictions = model.predict(X_test)

from sklearn.metrics import mean_absolute_error
print('MAE:', mean_absolute_error(predictions, y_test))

from sklearn.metrics import mean_squared_error
print('MSE:', mean_squared_error(predictions, y_test))

from sklearn.metrics import r2_score
print('r2_score:', r2_score(y_test, predictions))
```
